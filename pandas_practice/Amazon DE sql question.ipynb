A satellite television provider XYZ is looking forward to your help in retrieving some critical information from its data warehouse. The satellite television provider provides television channels bundled in packages to its subscribers.
The below diagram explains the key data tables and the relation-ship between them:
[Don't have diag]
- D_SUBSCRIBER is a dimension table that has details of subscribers (customers).
- D_PACKAGES is a dimension table that has details of various package offered by the provider
- F_SUBSCRIPTIONS is a fact table that stores package subscriptions by each subscriber. A subscriber can have multiple records in this table during his lifetime
---

Please note the following business rules
- There are 2 types of packages. BASE pack and ADDON pack (which is mentioned in PACKAGE_TYPE attribute)
- Subscriber will have at least 1 BASE pack. They can have 0 or more ADDON packs
- A package is considered active for a subscriber if PACKAGE_SUB_END_DT in F_SUBSCRIPTIONS is NULL. Otherwise the package is not active for the subscriber
- A subscriber is considered to be active if there is a BASE pack linked to him with NULL PACKAGE_SUB_END_DT. Otherwise the subscriber is considered to be inactive.
- Each package has a PRICE_PER_DAY which is the fee for the package for a single day usage. A subscriber’s monthly fee is the PRICE_PER_DAY of each of his active packages multiplied by No. of active days in that month that he has been on that package. The fee is considered as revenue to the business.

```python
%%sql
/*
The marketing team wants to run a campaign to bring back subscribers who are no longer active. Write a query to pull out 
subscribers who are no longer active?
*/
select D_subscriber_id from test.F_subscriptions s
join test.D_packages p on s.D_package_id = p.D_package_id
where p.package_type = 'Base'
and s.package_end_dt is not null
/*
group by D_subscriber_id
having sum(case when s.package_end_dt is null then 1 else 0 end)=0
*/
```

```python
%%sql
/*
The products team wants to understand subscribers who have at least 2 Add on Packages. Write a query to pull out subscribers 
who have at least 2 active ADDON pack?
*/
with active_subscribers as(
select D_subscriber_id as active_id from test.F_subscriptions s
join test.D_packages p on s.D_package_id = p.D_package_id
where p.package_type = 'Base'
and s.package_end_dt is null)

select D_subscriber_id from test.F_subscriptions s
join active_subscribers a on s.D_subscriber_id=a.active_id
join test.D_packages p on s.D_package_id = p.D_package_id
where p.package_type='Addon'
group by D_subscriber_id having count(*)>=2
```

```python
%%sql
/*
The marketing team believes that people who subscribe to ‘English News’ package should also be needing 
‘English Entertainment’ package. Write a SQL to pull out subscribers who have active ‘English News’ package 
but do not have active ‘English Entertainment’ package
*/
select D_subscriber_id as active_id from test.F_subscriptions s
join test.D_packages p on s.D_package_id = p.D_package_id
where p.package_type = 'Addon'
and s.package_end_dt is null
group by 1 having sum(case when package_name='English News' then 1 else 0 end) >= 1
    and sum(case when package_name='English Entertainment' then 1 else 0 end) =0
```

```python
%%sql
/* The finance team wants to track revenue generated by each subscriber. */
select D_subscriber_id, sum(price_per_day * datediff(ifnull(package_end_dt, date(now())), package_start_dt)) as total_amount
from test.F_subscriptions s
join test.D_packages p on s.D_package_id = p.D_package_id
group by 1
```

```python
%%sql
/* Write a SQL to get total revenue generated by each subscriber in the year 2014 */

with revenue_for_year as(
select D_subscriber_id as subscriber_id, D_package_id as package_id,
       case
       when year(package_start_dt) = 2013 then package_start_dt else null end as start_date,
       case
       when year(package_end_dt) = 2013 then package_end_dt else null end as end_date
from
       test.F_subscriptions
    
    having start_date is not null
)
select subscriber_id, sum(price_per_day * datediff(ifnull(end_date, '2013-12-31'), start_date)) as amount
from revenue_for_year r
join test.D_packages p
on r.package_id=p.D_package_id
group by 1
```

```python
%%sql
/* Write a query to identify Top 3 base packages (in terms of revenue collected) for each month in 2014 */

with recursive date_loop
(min_date, max_date) as(
select '2013-01-01' as min_date, '2013-12-31' as max_date
    union all
select date_add(min_date, interval 1 day) as min_date, max_date
    from date_loop
    where min_date<max_date
),
revenue_for_year as(
select D_subscriber_id as subscriber_id, D_package_id as package_id,
       case
       when year(package_start_dt) = 2013 then package_start_dt else null end as start_date,
       case
       when year(package_end_dt) = 2013 then package_end_dt else null end as end_date
from
       test.F_subscriptions
    
    having start_date is not null
)
select case when start_date>=min_date and ifnull(end_date,'2013-12-31') <= max_date then min_date end as x
       from revenue_for_year, date_loop
```

```python
%%sql
/* Write a query to identify Top 3 base packages (in terms of revenue collected) for each month in 2014 */

with recursive date_loop
(subscriber_id, package_id, min_date, max_date) as(
select D_subscriber_id as subscriber_id, D_package_id as package_id,
    min(package_start_dt) as min_date, max(ifnull(package_end_dt, date(now()))) as max_date
    from test.F_subscriptions group by 1,2
    union all
select subscriber_id, package_id, date_add(min_date, interval 1 day) as min_date, max_date
    from date_loop
    where min_date<max_date
)
select subscriber_id, package_id, year(min_date) year, sum(price_per_day) as amount
from date_loop dl join
test.D_packages p on dl.package_id=p.D_package_id
group by subscriber_id, package_id, year
having year=2013
```
