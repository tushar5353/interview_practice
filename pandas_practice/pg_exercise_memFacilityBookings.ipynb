Q2. Produce a list of each member name, id, and their first booking after September 1st 2012. Order by member ID.

```python
%%sql

select m.memid, firstname, min(starttime) as firstbooking
from cd.members m join cd.bookings b
on m.memid=b.memid
where starttime>='2012-09-01 00:00:00'
group by 1,2
```
